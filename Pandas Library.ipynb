```python
#Session- 11
#Basic arithmetic operations for series object
# 1. Scalar/Vector operations
import pandas as pd
import numpy as np
s=pd.Series([10,20,30,40,50,pd.NA])
print(s)
print(s+10)
print(s-3)
print(s*5)
print(s/2)
print()
print()
# 2. Arithmetic operation between two series object
s1=pd.Series([10,20,30])    #index:0,1,2
s2=pd.Series([10,20,30])    #index:0,1,2
print(s1+s2)
print()
s1=pd.Series(data=[10,20,30,40,50],index=['A','B','C','D','E'])    
s2=pd.Series(data=[10,20,30,40,50],index=['C','D','E','F','G'])    
print(s1+s2)
#using fill_value parameter
print(s1.add(s2,fill_value=0))
print()
#NaN example
s1=pd.Series(data=[10,pd.NA],index=['A','Z'])    #index:0,1,2
s2=pd.Series(data=[10,20],index=['A','B'])    #index:0,1,2
print(s1.add(s2,fill_value=0))
print()

#Subtraction,multi... and division
s1=pd.Series(data=[10,20,30,40,50],index=['A','B','C','D','E'])    
s2=pd.Series(data=[10,20,30,40,50],index=['C','D','E','F','G'])  
print(s1.sub(s2,fill_value=0))
print(s1.mul(s2,fill_value=0))
print(s1.div(s2,fill_value=0))
print()

print('-------Cumulative Operations-------')
s=pd.Series([10,20,30,40,50])
print(s.sum())
print(s.cumsum())   #returns cumulative sum
s=pd.Series([1,2,3,4,5])
print(s.prod())
print(s.cumprod())    #returns cumulative product
print()
print('Min and Cummin')
print(s.min())
print(s.cummin())  #returns cumulative min value
print()
print('Max and cummax')
print(s.max())
print(s.cummax())    #returns the cumulative maximum value
 
```

```python
#Session - 31
import pandas as pd
enos=[100,200,300,400,500,600]
enames=['Sunny','Bunny','Chinny','Kaminyy','Sunny','Munny']
esals=[1000.0,3500.0,2300.0,5000.0,2430,6000]
eaddrs=['Hyderabad','Pune','Hyderabad','Hyderabad','Delhi','Kolkata']
df=pd.DataFrame({'Eno':enos,'Ename':enames,'Esal':esals,'Eaddr':eaddrs})
print(df)
print()
#To print all hyderabad based employees
#df.query("Eaddr == 'Hyderabad'",inplace=True)
print(df.query("Eaddr == 'Hyderabad'"))
print()
print(df.query("Esal > 1500"))   #employess having salary >1500
print()
#find hyderabad based employee having salary >1500
print(df.query("Eaddr=='Hyderabad' and Esal > 1500"))
print()
#Select employee not based on hyderabad or having salary < 4000
print(df.query("Eaddr!='Hyderabad' and Esal < 4000"))
print()

#in and not in operator in query method
#To select employee based on hederabad and delhi
print(df.query("Eaddr in ['Hyderabad','Delhi']"))
print(df.query("Eaddr  not in ['Hyderabad','Delhi']"))
print()
#############################################
df1=pd.DataFrame({'A':range(1,6),'B':range(10,0,-2),'C C':range(10,5,-1)})
print(df1)
print()
print(df1.query('A < B'))   #Select elements where A< B
print()
print(df1.query('B == `C C`'))   #Select elements where B == C C
print()
print('-------------------apply() method----------------')
#Double salary of all employees
df['Esal'] = df['Esal'].apply(lambda x: 2*x)
print(df)
print()
#Ex. Double only Hyderabad based employee salary
def double_salary(row):
  eaddr=row[3]
  if eaddr =='Hyderabad':
    row[2] *=2
  return row
print(df.apply(double_salary,axis='columns'))
print()
#Ex. Add Designation column to the dataframe........
def find_designation(row):
  esal = row[2]
  if esal <= 2000:
    return 'Junior Employee'
  elif esal <= 5000:
    return 'Senior Employee'
  else:
    return 'Seniormost Employee'
df['Designation'] = df.apply(find_designation,axis='columns')
print(df)
  
```

```python
#Session - 33
#How to filter Dataframe with where() method 
import pandas as pd
enos=[100,200,300,400,500,600]
enames=['Sunny','Bunny','Chinny','Kaminyy','Sunny','Munny']
esals=[1000.0,3500.0,2300.0,5000.0,3500,6000]
eaddrs=['Hyderabad','Pune','Hyderabad','Hyderabad','Delhi','Kolkata']
df=pd.DataFrame({'Eno':enos,'Ename':enames,'Esal':esals,'Eaddr':eaddrs})
print(df)
print()
c=df['Eaddr']=='Hyderabad'
print(df.where(c))
print()
#Repalce salary as not useful where Eaddr is not Hyderabad
c=df['Eaddr']=='Hyderabad'
print(df.where(c,other='Not Useful'))
print()
#Replace salary with 'Less Salary' where salary <2700 ??
df['Esal'] = df['Esal'].where(lambda x: x>2700,other='Less Salary')
print(df)
print()
#Filter all values where esal<2500 and eaddr is 'Hyderabad'.
#c1=df['Esal'] < 2500
c2=df['Eaddr'] =='Hyderabad'
print(df.where(c2))
```

```python
#Session - 34
import pandas as pd
enos=[100,200,300,400]
enames=['Sunny leone','Bunny sharma','Chinny chaman','Kamini gupta']
esals=[10000,35040,23000,50000]
eaddrs=['Hyderabad','Pune','Hyderabad','Hyderabad']
df=pd.DataFrame({'Eno':enos,'Ename':enames,'Esal':esals,'Eaddr':eaddrs})
print(df)
df['Ename']=df['Ename'].str.upper()   #Make name letters uppercase
print(df)
print()
df['Ename']=df['Ename'].str.title()   #Make name letters titlecase
print(df)
df['Ename']=df['Ename'].str.strip().str.replace(' ','_')
print(df)
print()
#Add new column 'number of characters' containing length of ename values
df['Number of characters']=df['Ename'].str.len()
print(df)
```
